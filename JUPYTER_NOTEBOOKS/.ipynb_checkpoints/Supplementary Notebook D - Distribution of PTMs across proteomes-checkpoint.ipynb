### Description
This notebook depicts the distributions of post-translational modifications (PTMs) across the mouse, human and CHO proteomes for comparison purposes

```python
# Load necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Load PSIMs

```python
humanPSIM = pd.read_csv('RECON2s_python/PSIM_HUMAN.tab',sep='\t')
mousePSIM = pd.read_csv('iMM1685s_python/PSIM_MOUSE.tab',sep='\t')
choPSIM = pd.read_csv('iCHO2048s_python/PSIM_CHO.tab',sep='\t')
```

# Plot proportion of proteins with signal peptides across the three proteomes

```python
plt.figure(figsize=(10,10))
labels = 'Signal Peptide', 'No signal peptide'
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # explode 1st slice

# Plot Human
sp = len(humanPSIM[humanPSIM['SP']==1])
nsp = len(humanPSIM) - sp
sizes = [sp,nsp]
plt.subplot(3,1,1)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Human')

# Plot Mouse
sp = len(mousePSIM[mousePSIM['SP']==1])
nsp = len(mousePSIM) - sp
sizes = [sp,nsp]
plt.subplot(3,1,2)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Mouse')

# Plot CHO
sp = len(choPSIM[choPSIM['SP']==1])
nsp = len(choPSIM) - sp
sizes = [sp,nsp]
plt.subplot(3,1,3)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('CHO')
```

# Plot histograms of disulfide bonds (Log2)

```python
psims = [humanPSIM,mousePSIM,choPSIM]
labels = 'Human','Mouse','CHO'
colors = ['yellowgreen', 'lightcoral', 'lightskyblue']
plt.figure(figsize=(12,9))
for i in range(3):
    plt.subplot(3,1,i+1)
    x = np.array(psims[i]['DSB'])
    plt.hist(np.log2(x[x>0]),bins=50,color=colors[i],label=labels[i])
    plt.legend()
    plt.ylabel('Counts')
plt.xlabel('Log2 (Number of DSB)')
```
